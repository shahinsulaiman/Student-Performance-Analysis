```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")

# --- EDA ---
print("\n--- Dataset Overview ---")
print(Student_copy.info())
print("\n--- Descriptive Statistics ---")
print(Student_copy.describe(include="all"))
```

```python
# Check for missing values
print("\n--- Missing Values ---")
print(Student_copy.isnull().sum())
```

```python
# Numerical and categorical columns
numerical_cols = Student_copy.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = Student_copy.select_dtypes(include=["object"]).columns
print("\nNumerical Columns:", numerical_cols)
print("\nCategorical Columns:", categorical_cols)
```

```python
# --- Visualizations ---
# Distribution of numerical columns
Student_copy[numerical_cols].hist(figsize=(12, 10), bins=15, edgecolor="black")
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```python
# Count plots for categorical columns
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(Student_copy[col], order=Student_copy[col].value_counts().index)
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# --- Statistical Analysis ---
corr_matrix = Student_copy.corr(numeric_only=True)
print(corr_matrix)
```

```python
# Heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Grouped statistics for categorical variables
categorical_cols = Student_copy.select_dtypes(include=["object"]).columns
for col in categorical_cols:
    print(f"\nGrouped Summary for {col}:")
    print(Student_copy.groupby(col)["Exam_Score"].describe())
```
